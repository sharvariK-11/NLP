```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 8,16
```

```python
reader = easyocr.Reader(['en'])
```

```python
Image("/content/drive/MyDrive/ocr final.png")
```

```python
output = reader.readtext('/content/drive/MyDrive/ocr final.png')
```

```python
output
```

3 parts = coordinates , actual text , confidence
here, default language = hindi

```python
cord=output[-1][0]
```

```python
x_min,y_min=[min(idx) for idx in zip(*cord)]
```

```python
x_max,y_max=[max(idx) for idx in zip(*cord)]
```

```python
image=cv2.imread('/content/drive/MyDrive/ocr final.png')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```
